```python
import pandas as pd
import pandas as pd
from IPython.display import display, Image
import matplotlib.pyplot as plt
import pickle
```

```python
def load_results(filename):
    df = pd.read_parquet(filename)
    # Unpickle the numpy array
    df['original_image'] = df['original_image'].apply(pickle.loads)
    return df
```

```python
# Load the results
df = load_results("flagged_caroline_latin.parquet")
df
```

```python

# Filter to the images that have a .5 or higher CER
cer_df = df.loc[df["cer"] > .5]
cer_df
```

```python
# Display the DataFrame contents and images
for index, row in cer_df.iterrows():
    print("Text:", row['text'])
    print("Predicted Text:", row['predicted_text'])
    print("CER:", row['cer'])

    # Display the image from numpy array
    plt.figure(figsize=(5, 5))  # Adjust the size as needed
    if len(row['original_image'].shape) == 3 and row['original_image'].shape[2] == 3:
        plt.imshow(row['original_image'])
    else:
        plt.imshow(row['original_image'], cmap='gray')  # For grayscale images
    plt.axis('off')  # Hide axes
    plt.show()
    print("\n---\n")
```
